```
df_total_acc_x = pd.read_csv(dataset_path + 'test/Inertial Signals/total_acc_x_test.txt', header=None, delim_whitespace=True)
df_total_acc_y = pd.read_csv(dataset_path + 'test/Inertial Signals/total_acc_y_test.txt', header=None, delim_whitespace=True)
df_total_acc_z = pd.read_csv(dataset_path + 'test/Inertial Signals/total_acc_z_test.txt', header=None, delim_whitespace=True)
```

```
total_acc_x = df_total_acc_x.values.flatten()
total_acc_y = df_total_acc_y.values.flatten()
total_acc_z = df_total_acc_z.values.flatten()
```

```
i, x = 0, []
while len(x) != len(total_acc_x):
  x.append(i)
  i += 2.56
len(x), len(total_acc_x)
```

```
fig, ax = plt.subplots(nrows=3, ncols=1, figsize=(25, 8))
ax[0].plot(x, total_acc_x) 
ax[1].plot(x, total_acc_y)
ax[2].plot(x, total_acc_z)
ax[0].set_title('total acc x')
ax[1].set_title('total acc y')
ax[2].set_title('total acc z')
```
